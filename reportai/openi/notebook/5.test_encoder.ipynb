```python
import os
os.environ['CUDA_VISIBLE_DEVICES']='7'
import sys
import torch
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from glob import glob
```

```python
infile = './xray_data/features/CXR3139_IM-1476-2001.npy'
featrue_map = np.load(infile)
```

```python
in_feature = torch.from_numpy(featrue_map)
in_feature = torch.unsqueeze(in_feature, 0)
in_feature.shape
```

```python
class EncoderCNN(nn.Module):
    def __init__(self, embed_size):
        """Load the pretrained ResNet-152 and replace top fc layer."""
        super(EncoderCNN, self).__init__()
        self.conv1 = nn.Conv2d(1024, 1024, kernel_size=1)
        self.bn1 = nn.BatchNorm2d(1024)
        self.relu1 = nn.ReLU(inplace=True)
        self.pooling = nn.AdaptiveAvgPool2d((1, 1))
        self.linear = nn.Linear(1024, embed_size)
        self.bn = nn.BatchNorm1d(embed_size)
        self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self, in_features):
        """Extract feature vectors from input images."""        
        x = self.conv1(in_features)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.pooling(x)
        x = torch.flatten(x, 1)
        x = self.linear(x)
        print(x.shape)
#         features = torch.nn.BatchNorm1d(1024)(x)
        features = self.sigmoid(x)
        return features
```

```python
embedding_size = 1024
encoder = EncoderCNN(embedding_size)
```

```python
out_features = encoder(in_feature)
```
